# 📊 Модуль: Основы статистики (Введение)
## Урок 1.2: Генеральная совокупность и Выборка

> **Главная цель:** Изучить малую группу объектов так, чтобы сделать верный вывод обо всех сразу.

### 🏛 1. Основные понятия
*   **Генеральная совокупность (Population)** — «Абсолютно ВСЕ». Весь океан объектов, которые нас интересуют (например, все пользователи Python в мире).
*   **Выборка (Sample)** — «Те, кто под рукой». Часть океана, которую мы реально исследуем (например, 500 человек из чата в Telegram).

---

### ⚖️ 2. Параметры vs Статистики

| Тип | Где измеряем? | Обозначение | Суть |
| :--- | :--- | :--- | :--- |
| **Параметр** | Генеральная совокупность | Греческие буквы ($\mu$, $\sigma$) | Истинное значение (обычно неизвестно) |
| **Статистика** | Выборка | Латинские буквы ($\bar{x}$, $sd$) | То, что мы вычислили в PyCharm |

---

### 🎯 3. Репрезентативность (Качество данных)
Чтобы вывод был честным, выборка должна быть **репрезентативной** (мини-копия оригинала).

**Как этого добиться?**
*   **Простая случайная выборка (Simple Random Sample):** Каждый объект имеет равный шанс попасть «под микроскоп».
*   **Стратифицированная выборка:** Делим на группы (например, по возрасту) и берем из каждой группы пропорционально.

---

### 🧠 Связь с AGI (Путь к нейросетям)
В машинном обучении твоя таблица данных — это **выборка**.
*   Если ты учишь автопилот Tesla ездить только в солнечную погоду (**смещенная выборка**), он попадет в аварию в дождь.
*   **AGI** требует максимально репрезентативных данных со всего мира, чтобы «понимать» реальность целиком.

### 🌳 Классификация переменных

#### 1. Количественные (Quantitative) — "Сколько?"
Это числовые значения, которые можно складывать и вычитать.
*   **Непрерывные (Continuous):** Принимают любое значение на шкале (Рост: 180.5 см, Вес: 70.2 кг).
*   **Дискретные (Discrete):** Только целые числа, «счётные» объекты (Количество детей: 3, Количество машин: 1).

#### 2. Качественные (Qualitative) — "Какой?"
Это категории, свойства или признаки объектов.
*   **Номинативные (Nominal):** Просто названия или метки. Между ними **нет порядка** (Пол: М/Ж, Цвет волос, Номер школы).
*   **Ранговые (Ordinal):** Категории, которые можно **расставить по порядку**, но мы не знаем точного расстояния между ними (Место в забеге: 1, 2, 3; Уровень боли: низкий/средний/высокий).

# 🎯 Урок 1.4: Меры центральной тенденции (на примерах)
> **Задача:** Описать группу данных одним «типичным» числом.

### 📊 Пример данных: Зарплаты в маленьком офисе (в тыс. руб.)
Представим 5 сотрудников: `20, 20, 30, 40, 490` (последний — директор).

---

### 1. 🏠 Мода (Mode) — «Самый популярный»
*   **Что это:** Значение, которое встречается чаще всего.
*   **В нашем примере:** Это **20**, так как оно встречается дважды, а остальные — по одному разу.
*   **Гистограмма:** Самый высокий столбик на графике — это и есть мода.
*   **Применение:** Идеально для номинативных данных (например, «какой цвет глаз самый частый?»).

### 2. 📏 Медиана (Median) — «Золотая середина»
*   **Что это:** Число, которое стоит ровно посередине в **отсортированном** списке.
*   **В нашем примере:** `20, 20, [30], 40, 490`. Медиана = **30**.
*   **Сила:** Она проигнорировала огромную зарплату директора (490) и показала реальный центр. 50% людей получают меньше 30, 50% — больше.

### 3. 🧮 Среднее значение (Mean) — «Общий котел»
*   **Что это:** Сумма всех чисел / Количество сотрудников.
*   **В нашем примере:** $(20+20+30+40+490) / 5 = 600 / 5 = \mathbf{120}$.
*   **Проблема:** Средняя зарплата 120к, хотя почти все получают меньше 40к. Директор «испортил» среднее своими выбросами.

### 📝 Свойства среднего арифметического
Среднее значение обладает уникальными математическими особенностями, которые важно знать для анализа данных:

1. **Сумма отклонений всегда равна нулю**
   Если из каждого числа выборки вычесть среднее арифметическое и сложить результаты, мы всегда получим **0**.
   *   **Формула:** $\sum(x_i - \bar{x}) = 0$
   *   **Пример:** Для ряда `[2, 4]` среднее = 3. Отклонения: $(2-3)=-1$ и $(4-3)=1$. Итог: $-1 + 1 = 0$.
   *   *Суть:* Среднее — это идеальный "балансир" (точка равновесия) данных.

2. **Прибавление константы**
   Если к каждому числу в наборе прибавить число $C$, то и среднее увеличится на это же число $C$.
   *   **Формула:** $M(X + C) = M(X) + C$
   *   **Пример:** Зарплаты `[20, 30]`, среднее = 25. Всем подняли з/п на 5к: `[25, 35]`, новое среднее = 30.

3. **Умножение на константу**
   Если каждое число в наборе умножить на число $C$, то среднее увеличится в $C$ раз.
   *   **Формула:** $M(X \cdot C) = M(X) \cdot C$
   *   **Пример:** Те же зарплаты `[20, 30]`, среднее = 25. З/п увеличили в 2 раза: `[40, 60]`, новое среднее = 50.

4. **Минимизация квадратов отклонений**
   Сумма квадратов отклонений от среднего арифметического **минимальна** по сравнению с любым другим числом.
   *   **Формула:** $\sum(x_i - \bar{x})^2 = min$
   *   **Зачем это знать:** Это свойство лежит в основе метода наименьших квадратов (МНК) в регрессиях.

5. **Чувствительность к выбросам**
   Как видно из примера с директором, даже одно экстремально большое или маленькое значение "тянет" среднее на себя, искажая реальную картину для большинства.

---

### 📈 Как это выглядит на гистограмме (Распределения)

1. **Симметричное («Колокол»):** Среднее, Медиана и Мода находятся в одной точке — в самом центре. Это идеал для статистики.
2. **Положительная асимметрия (хвост вправо, как с зарплатами):**
   * Мода — на вершине (слева).
   * Медиана — чуть правее.
   * Среднее — утянуто далеко вправо "хвостом" (директором).

---

### 🧠 Сила для AGI и Кода:
* **В Pandas:** `df['salary'].median()` — спасет твою модель от "вранья" средних чисел.
* **В Пути к ИИ:** Если ты учишь нейросеть распознавать котиков, а на 1 фото из 100 попал слон — это выброс. Среднее значение пикселей "сломается", а медиана — нет.


# 📊 Тема 1.5: Меры изменчивости (Variability)

Меры изменчивости показывают, насколько данные "разбросаны" вокруг среднего.

---

### 1. Размах (Range)
Самая простая мера: разница между максимумом и минимумом.
* **Формула:** $R = X_{max} - X_{min}$
* **Суть:** Показывает только границы данных.
* **Минус:** Чувствителен к выбросам.

**🛠 Мини-пример:**
Данные: `[10, 15, 20, 100]`
Размах: $100 - 10 = 90$
*Если убрать выброс (100), размах упадет до 10.*

---

### 2. Дисперсия (Variance)
Средний квадрат отклонений от среднего арифметического.
* **Выборка:** $sd^2 = \frac{\sum (x_i - \bar{x})^2}{n - 1}$
* **Совокупность:** $\sigma^2 = \frac{\sum (x_i - \bar{x})^2}{n}$

**🛠 Мини-пример:**
Данные: `[2, 4]`, среднее = 3.
1. Отклонения: $(2-3)=-1$ и $(4-3)=1$
2. Квадраты: $(-1)^2=1$ и $1^2=1$
3. Дисперсия выборки: $(1+1) / (2-1) = 2$

---

### 3. Стандартное отклонение (Standard Deviation)
Корень из дисперсии. Возвращает нас к исходным единицам измерения.
* **Формула:** $sd = \sqrt{sd^2}$

**🛠 Мини-пример:**
Если дисперсия роста $sd^2 = 25 см^2$, то:
Стандартное отклонение $sd = 5 см$.
*Это значит, что в среднем данные отклоняются от центра на 5 см.*

---

### 4. Свойства дисперсии
Основные правила изменения разброса при модификации данных.


| Свойство | Формула | Мини-пример |
| :--- | :--- | :--- |
| **Константа** | $D(C) = 0$ | Для `[5, 5, 5]` дисперсия равна 0. |
| **Прибавление** | $D(X+C) = D(X)$ | Прибавили всем +10 баллов — разброс (расстояние между точками) не изменился. |
| **Умножение** | $D(X \cdot C) = D(X) \cdot C^2$ | Увеличили данные в 3 раза — дисперсия выросла в $3^2=9$ раз. |

> **NB:** При умножении на $C$ стандартное отклонение растет в $C$ раз, а дисперсия — в $C^2$.


### Реальный пример работы СКО

---


**Среднеквадратическое отклонение (СКО)** — это число, которое показывает, насколько сильно данные **отклоняются** или **разбросаны** относительно своего среднего значения.

Проще говоря:
*   **Малое СКО** означает, что все значения сгруппированы tightly вокруг среднего (как кучка снайперских пуль на мишени). Данные предсказуемые и стабильные.
*   **Большое СКО** означает, что значения разбросаны широко (как пули из дробовика). Данные изменчивые и непредсказуемые.

**Где это применяется?** В финансах (риск инвестиций), в контроле качества (стабильность производства), в метеорологии (колебания температуры), в науке и т.д.

---

### Простой пример: Два кафе

Представьте, что вы выбираете, в какое кафе пойти пить кофе. Вам важно, чтобы кофе был не слишком горячий. Вы целую неделю измеряете температуру кофе в двух кафе сразу после приготовления.

Средняя температура почти одинаковая! Если бы мы смотрели только на среднее, оба кафе были бы одинаково хороши.

**Но посмотрите на цифры!** В первом кафе температура всегда около 88°. Во втором — она скачет от очень холодного (70°) до очень горячего (95°).

*   **СКО для "Стабильности": ~1.4°C**
*   **СКО для "Разброса": ~8.9°C**

**Шаг 3: Анализируем результат.**

*   В **Кафе "Стабильность"** очень **малое СКО** (~1.4). Это значит, что температура кофе очень стабильна и предсказуема. Вы всегда знаете, что получите кофе температуры около 88°.

# 📊 Тема 1.6: Квантили, Квартили и Box Plot

В статистике и машинном обучении важно знать не только среднее значение, но и то, как данные распределены по "группам" и где находятся аномалии.

---

## 1.1 Квартили (Quartiles)
**Квартили** — это три точки (значения), которые делят упорядоченный набор данных на **четыре равные части** (по 25% данных в каждой).

*   **Q1 (Первый квартиль):** 25-й процентиль. Отсекает 25% самых маленьких значений.
*   **Q2 (Второй квартиль):** 50-й процентиль. Это **Медиана**. Делит выборку пополам.
*   **Q3 (Третий квартиль):** 75-й процентиль. Отсекает 75% данных (или 25% самых больших).

> **Пример:** В ряду `[1, 2, 3, 4, 5, 6, 7, 8]` медиана (Q2) = 4.5, Q1 = 2.5, Q3 = 6.5.

---

## 1.2 Квантили (Quantiles)
**Квантиль** — это более общее понятие. Это значение, которое заданный набор данных не превышает с определенной вероятностью $P$.

*   Квартили — это частный случай квантилей (с шагом 0.25).
*   Процентили — квантили с шагом 0.01.
*   **Связь:** Квантиль 0.5 = 2-й квартиль = Медиана.

---

## 2. Понятие Box plot («Ящик с усами»)
**Box plot** — это график, который визуализирует распределение данных на основе пяти показателей (Five-number summary).

### Из чего состоит график:
1.  **Нижняя граница (Ус):** $Q1 - 1.5 \cdot IQR$ (минимальное нормальное значение).
2.  **Нижняя стенка ящика:** Первый квартиль ($Q1$).
3.  **Линия внутри ящика:** Медиана ($Q2$).
4.  **Верхняя стенка ящика:** Третий квартиль ($Q3$).
5.  **Верхняя граница (Ус):** $Q3 + 1.5 \cdot IQR$.
6.  **Точки (Outliers):** Значения, выходящие за пределы "усов" (аномалии).

*Примечание: $IQR$ (Интерквартильный размах) = $Q3 - Q1$.*

### Зачем это в Machine Learning (ML):
*   **Детекция выбросов:** Быстрый поиск аномалий, которые могут "испортить" обучение модели.
*   **Сравнение признаков:** Позволяет визуально сравнить распределение разных категорий (например, зарплаты в зависимости от грейда).
*   **Проверка симметрии:** Помогает понять, есть ли у данных перекос (skewness) в одну из сторон.


````python
import math

def calculate_quartiles(data):
    # Данные обязательно должны быть отсортированы
    sorted_data = sorted(data)
    n = len(sorted_data)

    def get_quantile(p):
        """Находит значение для квантиля p (от 0 до 1)"""
        pos = p * (n - 1)
        base = math.floor(pos)
        rest = pos - base
        if (base + 1) < n:
            return sorted_data[base] + rest * (sorted_data[base + 1] - sorted_data[base])
        else:
            return sorted_data[base]

    q1 = get_quantile(0.25)
    q2 = get_quantile(0.50) # Она же Медиана
    q3 = get_quantile(0.75)
    iqr = q3 - q1

    # Границы для Box Plot (усы)
    lower_whisker = q1 - 1.5 * iqr
    upper_whisker = q3 + 1.5 * iqr

    print(f"Отсортированные данные: {sorted_data}")
    print(f"Q1: {q1}")
    print(f"Q2 (Медиана): {q2}")
    print(f"Q3: {q3}")
    print(f"IQR (Межквартильный размах): {iqr}")
    print(f"Нижний ус: {lower_whisker}")
    print(f"Верхний ус: {upper_whisker}")

    # Поиск выбросов
    outliers = [x for x in sorted_data if x < lower_whisker or x > upper_whisker]
    print(f"Выбросы: {outliers if outliers else 'Нет'}")

# Тестовый набор данных с явным выбросом (100)
my_data = [1, 5, 2, 7, 1, 9, 3, 8, 5, 100]
calculate_quartiles(my_data)
````

# 🔔 Тема 1.7: Нормальное распределение и Z-преобразование

### 1. Понятие нормального распределения
**Нормальное распределение** (Гауссово) — это распределение данных в форме симметричного колокола.

*   **Ключевые свойства:**
    *   График симметричен относительно центра.
    *   **Среднее = Медиана = Мода**.
*   **Параметры:**
    *   $\mu$ (Мю) — среднее значение (центр).
    *   $\sigma$ (Сигма) — стандартное отклонение (ширина "разброса").

---

### 📏 2. Правило «Трех сигм» (Грубая оценка)
Используется для быстрой прикидки процента данных в уме:
*   $M \pm 1\sigma \approx \mathbf{68\%}$ данных (в одном хвосте остается ~16%).
*   $M \pm 2\sigma \approx \mathbf{95\%}$ данных (в одном хвосте остается ~2.5%).
*   $M \pm 3\sigma \approx \mathbf{99.7\%}$ данных (практически все наблюдения).

---

### ⚖️ 3. Z-преобразование (Точная настройка)
Процесс перевода любого значения в стандартную шкалу, где **Среднее = 0**, а **СКО = 1**. Это позволяет сравнивать данные из разных групп.

**Формула Z-оценки:**
$$Z = \frac{x_i - \bar{x}}{sd}$$

*   **Z-score** показывает, на сколько стандартных отклонений число отстоит от среднего.
*   **Z > 0**: значение выше среднего.
*   **Z < 0**: значение ниже среднего.

---

### 📑 4. Как читать результаты (Таблицы и Вероятности)
Z-оценка — это "адрес" точки, а значение в таблице (или калькуляторе) — это **площадь (процент данных)**.

*   **Хвост (Tail):** Процент данных, которые оказались больше (правый) или меньше (левый) нашего значения.
*   **Логика 50%:** Так как колокол симметричен, каждая его половина — это 0.5 (50%).
    *   Если хвост за точкой $Z=1$ равен **0.1587**, то "кусок" от среднего до этой точки равен $0.5 - 0.1587 = \mathbf{0.3413}$ (или 34%).

---

### 🔗 5. Инструментарий
Для точных расчетов используй интерактивный калькулятор:
👉 [Distribution Calculator](https://shinyapps.io)

> **Инструкция:** Выстави `Mean = 0`, `SD = 1` и вводи рассчитанный **Z-score** в поле **a**. Программа покажет точный процент наблюдений ($P$).

---

### 🛠 Пример расчета:
Дано: $M=100$, $SD=15$. Найти процент людей с $IQ > 125$.
1. **Считаем Z:** $(125 - 100) / 15 = \mathbf{1.67}$.
2. **Смотрим в калькулятор:** Для $Z=1.67$ (Upper Tail) получаем $P = 0.0475$.
3. **Итог:** Примерно **4.75%** людей имеют IQ выше 125.


# Тема 1.8: Центральная предельная теорема (ЦПТ)
> **Суть:** Если мы берем много выборок из любой (даже самой "кривой") популяции, то их **средние значения** всегда выстроятся в красивый "колокол" (нормальное распределение).

### 🗝️ Ключевые выводы урока:
1. **Среднее выборочных средних** ($\mu_{\bar{x}}$) примерно равно **среднему всей популяции** ($\mu$).
2. **Стандартная ошибка среднего (Standard Error, SE)** — это стандартное отклонение этого самого "колокола" из средних значений. Она показывает, насколько точно выборочное среднее оценивает среднее популяции.
   $$SE = \frac{\sigma}{\sqrt{n}}$$
   *где $\sigma$ — стандартное отклонение популяции, $n$ — размер выборки.*
3. **Зависимость точности:**
   * Чем **больше** размер выборки ($n$), тем **меньше** ошибка ($SE$) и тем "узкий" и точный наш колокол.
   * Чем **больше** разброс в исходных данных ($\sigma$), тем **больше** ошибка ($SE$).

### 💡 Почему это база для ML?
В машинном обучении мы почти никогда не видим все данные мира (генеральную совокупность). Мы работаем с выборкой. ЦПТ позволяет нам понимать, насколько мы можем доверять средним значениям в наших данных.


````python
import random
import matplotlib.pyplot as plt

# --- ШАГ 1: Создаем "плохие" исходные данные ---
# Представь, что это количество лайков. Большинство 0-10, но есть редкие выбросы.
# Это абсолютно НЕ нормальное распределение.
population = [random.randint(0, 10) for _ in range(5000)] + [random.randint(50, 100) for _ in range(500)]

# --- ШАГ 2: Параметры эксперимента ---
n = 50              # Берем по 50 человек в каждой выборке
samples_count = 1000 # Повторяем это 1000 раз

sample_means = []   # Сюда будем складывать средние каждой выборки

# --- ШАГ 3: Магия ЦПТ (Цикл) ---
for i in range(samples_count):
    # Достаем случайные n элементов
    sample = random.sample(population, n)
    # Считаем их среднее и сохраняем
    mean_value = sum(sample) / n
    sample_means.append(mean_value)

# --- ШАГ 4: Визуализация ---
plt.figure(figsize=(10, 5))
plt.hist(sample_means, bins=30, color='lightgreen', edgecolor='black')
plt.title(f"Распределение средних значений (n={n})\nТеперь это 'Колокол'!")
plt.xlabel("Выборочное среднее")
plt.ylabel("Частота (как часто выпадало такое среднее)")
plt.show()

# Итоговая проверка
pop_mean = sum(population) / len(population)
clt_mean = sum(sample_means) / len(sample_means)

print(f"Среднее всей популяции: {pop_mean:.2f}")
print(f"Среднее из 'колокола' средних: {clt_mean:.2f}")
print("-" * 30)
print(f"Они почти совпали! Это и есть работа ЦПТ.")
````

---
### 📝 Практические задачи для закрепления

#### Задача 1: Расчет Standard Error (SE)
**Условие:** В генеральной совокупности стандартное отклонение $\sigma = 1000$. Мы извлекаем выборку размером $n = 100$. Чему равна стандартная ошибка среднего?
*   **Решение:** $SE = \frac{\sigma}{\sqrt{n}} = \frac{1000}{\sqrt{100}} = \frac{1000}{10} = 100$.
*   **Вывод:** Чем больше выборка, тем меньше ошибка. Чтобы уменьшить ошибку в 2 раза, нужно увеличить выборку в 4 раза.

#### Задача 2: Размер выборки и ЦПТ
**Вопрос:** Сработает ли "магия" ЦПТ (превращение в нормальный колокол), если размер выборки $n = 2$?
*   **Ответ:** Скорее всего, **нет**.
*   **Нюанс:** ЦПТ начинает работать эффективно при "достаточно большом" объеме данных. В статистике общепринятым порогом считается **$n \ge 30$**. Если выборка слишком мала, распределение средних будет повторять форму исходных (возможно, кривых) данных.

#### Задача 3: В чем разница между $\sigma$ и $SE$?
Это самый частый вопрос на собеседованиях:
*   **$\sigma$ (Standard Deviation):** Показывает разброс **индивидуальных значений** (насколько один человек отличается от другого).
*   **$SE$ (Standard Error):** Показывает разброс **средних значений** (насколько среднее одной выборки может отличаться от среднего другой выборки или от истинного среднего ГС).

> **Золотое правило:** $SE$ — это всегда стандартное отклонение для распределения выборочных средних.


# Тема 1.9: Доверительные интервалы (Confidence Intervals)
> **Суть:** Мы не можем знать точное среднее всей популяции ($\mu$), поэтому мы строим "коридор" вокруг нашего выборочного среднего ($\bar{x}$), в который с вероятностью 95% или 99% попадает истина.

### 📝 Основная формула:
$$\bar{x} \pm Z \cdot \frac{sd}{\sqrt{n}}$$

Где:
*   $\bar{x}$ — выборочное среднее (**Центр интервала**).
*   $Z$ — коэффициент уверенности (зависит от выбранного уровня доверия).
*   $sd / \sqrt{n}$ — стандартная ошибка (**SE**), которая определяет масштаб "шага" ошибки.

### 📊 Популярные значения Z:
*   **95% доверия** (ошибка 5%): $Z = 1.96$
*   **99% доверия** (ошибка 1%): $Z = 2.58$

### 🔍 Главные зависимости (Ширина интервала):
1.  **Размер выборки ($n$):** Увеличиваем $n$ в 4 раза — интервал становится **уже в 2 раза** (точность растет).
2.  **Разброс данных ($sd$):** Чем хаотичнее данные, тем **шире** интервал.
3.  **Уровень доверия:** Чем выше мы хотим быть уверены (99% вместо 95%), тем **шире** становится наш "забор".


## 💡 FAQ: Ответы на сложные вопросы

### 1. Откуда взялось число 1.96? (Z-шкала и "хвосты")
Это значение из **Z-преобразования**.
*   Мы хотим оставить 95% вероятности **в центре** распределения.
*   Это значит, что на два хвоста остается по 2.5% ошибки (итого 5%).
*   По таблице Z-значений именно отступ в **1.96 сигм** отсекает ровно 2.5% с каждого края.

### 2. Почему мы умножаем Z на SE?
*   **$SE$** — это "линейка" для средних значений. Она показывает, насколько в среднем ошибается среднее нашей выборки.
*   **$Z$** — это количество таких "линеек", которые мы откладываем в обе стороны, чтобы достичь нужной уверенности.
*   $\text{Отступ} = \text{Кол-во шагов} \times \text{Размер шага}$.

### 🚀 Продвинутые нюансы:
*   **Если мы не знаем $\sigma$ популяции:** При $n > 30$ мы имеем право использовать стандартное отклонение из нашей выборки ($sd$).
*   **Интервал и Ноль:** Если ДИ для разницы двух групп включает в себя **0**, значит статистически значимой разницы между ними нет (эффект может быть случайным).
*   **Компромисс:** Нельзя одновременно повысить и точность (узкий интервал), и уверенность (99%), не увеличивая размер выборки.


````python
import math

def calculate_ci(mean, sd, n, confidence=0.95):
    # 1. Находим SE
    se = sd / math.sqrt(n)

    # 2. Выбираем Z (основные значения для тестов)
    if confidence == 0.99:
        z = 2.58
    else:
        z = 1.96 # по умолчанию для 95%

    # 3. Считаем границы
    lower = mean - z * se
    upper = mean + z * se

    return lower, upper, z * se

# Пример из задачи: среднее 10, sd 5, n 100, доверие 99%
m, s, n_val = 10, 5, 100
low, high, margin = calculate_ci(m, s, n_val, confidence=0.99)

print(f"Результат для 99% ДИ:")
print(f"Интервал: [{low:.2f} ; {high:.2f}]")
print(f"Радиус (Margin of Error): {margin:.2f}")

# Эксперимент: как сжимается интервал при росте n
print("\nСравнение ширины при разном n (95% CI):")
for n_exp in [100, 400, 1600]:
    _, _, radius = calculate_ci(10, 5, n_exp)
    print(f"При n={n_exp:4} ширина интервала = {radius*2:.2f}")
````

### 💎 Разбор эталонной задачи (из теста Stepik)
**Условие:** $\bar{x} = 10, sd = 5, n = 100$. Построить доверительный интервал 99%.

**Пошаговое решение:**
1. **Находим размер шага (SE):**
   $$SE = \frac{sd}{\sqrt{n}} = \frac{5}{\sqrt{100}} = 0.5$$
2. **Выбираем уверенность (Z):**
   Для 99% мы берем $Z = 2.58$.
3. **Считаем радиус (Margin of Error):**
   $$Z \cdot SE = 2.58 \cdot 0.5 = 1.29$$
4. **Итоговый интервал:**
   * Нижняя граница: $10 - 1.29 = 8.71$
   * Верхняя граница: $10 + 1.29 = 11.29$

> **Запомни:** Радиус (1.29) — это максимально возможное отклонение нашего выборочного среднего от истинного, в котором мы уверены на 99%.


# Тема 1.10: Идея статистического вывода и P-value

> **Статистический вывод** — это процесс принятия решения о всей популяции на основе данных из одной выборки.

---

### 1. Фундамент: Гипотезы
Любая проверка начинается с двух утверждений:
*   **Нулевая гипотеза ($H_0$):** "Ничего не произошло". Эффекта нет, связи нет, всё — случайный шум.
*   **Альтернативная гипотеза ($H_1$):** "Эффект есть". Наше предположение верно.

### 2. Что такое P-value? (На понятном языке)
**P-value** — это вероятность получить наши данные (или еще более экстремальные), если мы допустим, что **на самом деле верна Нулевая гипотеза**.

*   **Маленький p-value (< 0.05):** "Событие слишком редкое для случайности". Значит, $H_0$ скорее всего неверна. **Отклоняем $H_0$**.
*   **Большой p-value (> 0.05):** "Такое часто случается само по себе". У нас недостаточно улик против $H_0$. **Не отклоняем $H_0$**.

---

### 🎮 Пример: Компьютерные игры и агрессия
*   **$H_0$:** Игры не влияют на количество драк.
*   **Эксперимент:** Геймеры дерутся в среднем чаще. Мы посчитали **p = 0.04**.
*   **Логика:** Если бы игры НЕ влияли, шанс получить такой разрыв случайно — всего 4%.
*   **Вердикт:** 4% < 5% $\implies$ Результат **статистически значим**. Мы отклоняем версию о случайности.

---

### 🚫 Ошибки статистического вывода
Мы всегда рискуем ошибиться, так как работаем с вероятностями:


| Ошибка | Название | Суть | Аналогия |
| :--- | :--- | :--- | :--- |
| **I рода** | Альфа ($\alpha$) | Отклонили ВЕРНУЮ $H_0$ | "Ложная тревога" (осудили невиновного) |
| **II рода** | Бета ($\beta$) | НЕ отклонили ЛОЖНУЮ $H_0$ | "Пропуск цели" (отпустили преступника) |

---

### 📏 Алгоритм решения задач (Кейс с баллами Stepik)
**Задача:** Ожидали $\mu=115$ баллов. Получили $\bar{x}=118$ при $n=144, sd=9$.

1. **Считаем стандартную ошибку (SE):**
   $$SE = \frac{sd}{\sqrt{n}} = \frac{9}{12} = 0.75$$
2. **Считаем Z-отклонение (на сколько шагов мы ушли от центра):**
   $$Z = \frac{\bar{x} - \mu}{SE} = \frac{118 - 115}{0.75} = \frac{3}{0.75} = 4$$
3. **Принимаем решение:**
   Так как $|Z| = 4$ (это намного больше 1.96), значит наш результат лежит в глубоком хвосте.
   **Вывод:** $p < 0.05$. Отклоняем нулевую гипотезу.

---

### 🔗 Связь с Доверительными Интервалами
Это два зеркальных способа проверки:
*   Если **тестируемое значение** (из $H_0$) **НЕ ПОПАДАЕТ** в 95% доверительный интервал $\implies$ **p < 0.05**.
*   Это значит, что "ожидание" находится за пределами нашего "забора доверия".

---

### ⚠️ Главные ловушки (НЕ делай так!):
*   $P$-value **НЕ** вероятность того, что гипотеза верна.
*   $P$-value **НЕ** показывает силу эффекта (разница может быть мизерной, но "значимой" на большой выборке).
*   $P > 0.05$ **НЕ** означает, что $H_0$ верна. Это значит лишь, что мы "не смогли её опровергнуть".


````python
import math
from scipy import stats

def check_hypothesis(mu, x_bar, sd, n):
    # 1. Считаем SE
    se = sd / math.sqrt(n)

    # 2. Считаем Z-значение
    z_score = (x_bar - mu) / se

    # 3. Находим p-value (двусторонний тест)
    p_value = stats.norm.sf(abs(z_score)) * 2

    print(f"Z-значение: {z_score:.2f}")
    print(f"P-value: {p_value:.5f}")

    if p_value < 0.05:
        print("Результат: p < 0.05. ОТКЛОНЯЕМ нулевую гипотезу (разница значима)")
    else:
        print("Результат: p > 0.05. НЕ ОТКЛОНЯЕМ нулевую гипотезу (разница случайна)")

# Данные задачи про баллы
check_hypothesis(mu=115, x_bar=118, sd=9, n=144)
````

# Тема 2.1: Т-распределение

## 1. Проблема малых выборок
Когда мы работаем с реальными данными, мы почти никогда не знаем стандартное отклонение генеральной совокупности ($\sigma$). Мы вынуждены оценивать его по выборке ($s$ или $sd$).

* Если выборка **большая** (обычно $n \ge 30$), используем **Z-критерий** (нормальное распределение).
* Если выборка **маленькая** ($n < 30$), оценка $sd$ становится менее точной. В таких случаях обычное нормальное распределение работает плохо, так как оно не учитывает дополнительную неопределённость.

## 2. Распределение Стьюдента (t-распределение)
Для решения этой проблемы Уильям Госсет (псевдоним *Стьюдент*) предложил **t-распределение**.

**Ключевые свойства:**
* **Форма:** похоже на нормальное (симметричный «колокол»), но имеет **более тяжёлые хвосты**. Это значит, что крайние значения (выбросы) встречаются в нём чаще, чем в нормальном.
* **Адаптивность:** чем меньше объём выборки, тем более «приплюснутым» становится колокол. С ростом объёма выборки t-распределение стремится к нормальному.

## 3. Понятие числа степеней свободы (Degrees of Freedom)
**Число степеней свободы ($df$)** — главный параметр t-распределения, определяющий его форму.

* **Формула для одного среднего:** $df = n - 1$
* **Суть:** при расчёте выборочного среднего мы накладываем одно ограничение на данные. Из $n$ наблюдений свободно варьироваться могут только $n-1$.

### 3.1 Почему именно $n-1$? (Интуиция)

#### А. Пример «Семь футболок» (свобода выбора)
Представьте, что у вас есть **7 футболок** на неделю.
* С понедельника по субботу вы каждый день вольны выбирать любую из оставшихся.
* Но в воскресенье у вас **нет выбора** — вы обязаны надеть ту единственную, что осталась.
* **Итог:** из 7 дней у вас было только **6 дней свободы** выбора.
  $$df = 7 - 1 = 6$$

#### Б. Математический пример (заложник среднего)
Представьте, что среднее трёх чисел равно 10. Значит, их сумма обязана быть 30.
1. Первое число может быть любым (например, 5).
2. Второе число может быть любым (например, 15).
3. Третье число автоматически должно быть $30 - 5 - 15 = 10$ — у него нет свободы.
* **Итог:** $df = 3 - 1 = 2$.

### Зачем это статистике?
Число степеней свободы ($df$) сообщает t-распределению, насколько «надёжна» наша выборка.
* Чем **меньше $df$**, тем более «расплывчатым» и широким становится график (распределение закладывает больший риск ошибки).
* Чем **больше $df$**, тем меньше мы «штрафуем» себя за использование выборочного среднего, и график стремится к нормальному.

## 4. Сравнение критических значений (95% доверительный интервал)
Чтобы «накрыть» 95% данных в центре, нужно отступить от среднего на разное количество стандартных ошибок (сигм).

| Распределение | Объём выборки ($n$) | $df$ | Критическое значение (кол-во сигм) |
|--------------|---------------------|------|-------------------------------------|
| Нормальное (Z) | $\infty$ | – | **1.96** |
| Стьюдента (t) | 30 | 29 | **2.04** |
| Стьюдента (t) | 5  | 4  | **2.78** |

**Вывод:** чем меньше выборка, тем больше должно быть критическое значение $t$, чтобы мы могли признать результат значимым.

## 5. Одновыборочный t-критерий (one-sample t-test)
Используется, когда мы хотим сравнить среднее нашей выборки ($\bar{x}$) с известным (гипотетическим) значением в генеральной совокупности ($\mu$).

**Формула t-значения:**
$$t = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}}$$
где $\frac{s}{\sqrt{n}}$ — стандартная ошибка среднего ($SE$).

**Логика:** мы смотрим, на сколько стандартных ошибок наше выборочное среднее отклонилось от предполагаемого $\mu$. Затем находим p-value для полученного $t$ при $df = n - 1$.

````python
from scipy import stats

# Находит t-критическое для 95% интервала (оставляем по 2.5% в хвостах)
t_crit = stats.t.ppf(0.975, df=19)
print(t_crit)

def calculate_p_value(t_value, df):
    """
    Вычисляет двустороннее p-value для t-распределения.
    t_value: полученное значение t-критерия
    df: число степеней свободы (n - 1)
    """
    # Используем модуль (abs), так как тест двусторонний (два хвоста)
    # stats.t.cdf(-abs(t)) находит площадь одного (левого) хвоста
    # Умножаем на 2, чтобы учесть оба хвоста
    p_value = 2 * stats.t.cdf(-abs(t_value), df)
    return p_value

# ПРИМЕР ИЗ КУРСА:
# Допустим, t = 2.5, n = 15 (значит df = 14)
t_val = 8.18
degrees_of_freedom = 198

p = calculate_p_value(t_val, degrees_of_freedom)

print(f"При t = {t_val} и df = {degrees_of_freedom}")
print(f"p-value (двустороннее): {p:.4f}")

if p < 0.05:
    print("Результат статистически значим (отвергаем H0)")
else:
    print("Результат не значим (не удалось отвергнуть H0)")

````

````python
import numpy as np
from scipy import stats

# Данные для мужчин
mean_m = 45
sd_m = 9
n_m = 100

# Данные для женщин
mean_w = 34
sd_w = 10
n_w = 100

# Стандартная ошибка разности средних (SE)
# SE = sqrt(sd1^2/n1 + sd2^2/n2)
se = np.sqrt((sd_m**2 / n_m) + (sd_w**2 / n_w))

# t-критерий
t_stat = (mean_m - mean_w) / se

# Степени свободы (df)
df = n_m + n_w - 2

# Двустороннее p-value
p_val = 2 * stats.t.cdf(-abs(t_stat), df)

print(f"{t_stat=}")
print(f"{p_val=}")
````

# Тема 2.2: Сравнение двух средних. t-критерий Стьюдента

## 1. Назначение метода
Используется для проверки гипотезы о равенстве средних значений в **двух независимых группах**.
*   **Пример**: Отличается ли средний рост мужчин и женщин?
*   **Переменные**:
    1. Одна качественная (группирующая) — строго 2 уровня (М/Ж, Тест/Контроль).
    2. Одна количественная (зависимая) — то, по чему считаем среднее.

## 2. Гипотезы
*   **$H_0$ (Нулевая)**: $\mu_1 = \mu_2$ (различий нет, средние в популяциях равны).
*   **$H_1$ (Альтернативная)**: $\mu_1 \neq \mu_2$ (различия статистически значимы).

## 3. Расчет t-значения
Формула показывает, сколько «стандартных ошибок» умещается в разности между средними:

$$t = \frac{\bar{x}_1 - \bar{x}_2}{SE}$$

Где $SE$ (стандартная ошибка разности средних):
$$SE = \sqrt{\frac{sd_1^2}{n_1} + \frac{sd_2^2}{n_2}}$$

## 4. Число степеней свободы ($df$)
Для двух независимых групп:
$$df = n_1 + n_2 - 2$$

## 5. Доверительный интервал для разности средних
Если интервал для разности $(\bar{x}_1 - \bar{x}_2)$ **не включает в себя 0**, значит:
1. Различия статистически значимы.
2. $p-value < 0.05$.
3. Мы отвергаем $H_0$.

## 6. Условия применения t-критерия
Для того чтобы t-тесту можно было доверять, должны соблюдаться:
1. **Нормальность распределения**: признак в обеих группах должен быть распределен нормально (особенно при $n < 30$).
2. **Гомогенность (равенство) дисперсий**: разброс данных в группах должен быть примерно одинаковым.
3. **Независимость**: наблюдения в группах не должны быть связаны друг с другом.

# Тема 2.2 (дополнение): Терминология и виды T-тестов

## 1. Основные понятия
*   **T-тест**: Статистический тест для определения значимости разницы между средними значениями двух групп.
    *   *Важно*: Предполагает **нормальное распределение** данных. Если это условие нарушено, результаты могут быть недостоверными.
*   **T-статистика (наблюдаемое значение t)**: Числовой результат теста. Это отношение разницы между средними к стандартной ошибке этой разницы.
    *   *Т-критерий* — более общий термин для метода, использующего t-распределение.
*   **Критическое значение T**: Пороговое значение из таблицы t-распределения. Оно зависит от:
    1. Уровня значимости $\alpha$ (обычно $0.05$).
    2. Числа степеней свободы $df$.
    *   Если $|T_{статистика}| > T_{критическое}$, различия статистически значимы.

## 2. Виды t-тестов

### А. Одновыборочный t-тест (One-sample t-test)
Сравнивает среднее выборки с заданным эталонным значением ($\mu_{предп}$).
**Формула:**
$$t = \frac{\bar{x} - \mu_{предп}}{\frac{s}{\sqrt{n}}}$$

### Б. Двухвыборочный независимый t-тест (Independent t-test)
Сравнивает средние двух разных, не связанных между собой групп.
**Формула:**
$$t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}$$

### В. Парный зависимый t-тест (Paired t-test)
Используется для связанных выборок, когда одна группа измеряется дважды (например, "до" и "после" воздействия).
*   Анализируется средняя разница между парами наблюдений.


# Тема 2.3: Проверка условий и "План Б" (Манн-Уитни)

Перед применением t-теста мы обязаны провести "диагностику" данных.

## 1. Проверка на нормальность: Тест Шапиро-Уилка
Этот тест — наш **"Одинокий охранник"** на входе в клуб "T-тест".

*   **Логика p-value здесь перевернута**:
    *   **$p > 0.05$ (Охранник спокоен)**: "Ваши данные выглядят нормально, проходите к t-тесту". Мы **не отклоняем** $H_0$.
    *   **$p < 0.05$ (Охранник заметил подвох)**: "Ваши данные кривые, t-тест здесь наврал бы". Мы **отклоняем** $H_0$ (нормальности нет).

## 2. Проблема выбросов: "Спортсмен-супермен"
Почему t-тест боится выбросов? Потому что он считает **среднее арифметическое**.

*   **Пример**: В группе 5 человек занимаются спортом 2 раза в неделю, и один "супермен" занимается **100 раз**.
*   **Результат**: Среднее взлетает до небес, но стандартное отклонение (разброс) становится гигантским.
*   **Итог**: T-тест "пугается" огромного разброса и выдает $p > 0.05$ (не значимо), даже если на самом деле группа активнее остальных. Он просто "слепнет" от одной яркой цифры.

## 3. U-критерий Манна-Уитни: Переход к рангам
Это наш **"Внедорожник"**, который едет по любым данным. Вместо средних значений он использует **ранги** (места в очереди).

*   **Суть**: Нам всё равно, сколько раз занимается "супермен" — 100 или 1 000 000. В очереди он просто будет стоять **последним** (получит самый большой ранг).
*   **Результат**: Выброс больше не раздувает ошибку. Тест видит реальную разницу в "позициях" между группами и выдает точное $p < 0.05$.

---

### Итоговая шпаргалка по принятию решений:


| Шаг | Инструмент | Результат | Что делать дальше? |
| :--- | :--- | :--- | :--- |
| **1. Диагноз** | **Shapiro-Wilk** | $p > 0.05$ | Все Ок, используем **t-тест** |
| | | $p < 0.05$ | Данные кривые, используем **Манна-Уитни** |
| **2. Поиск разницы** | **t-тест / Манн-Уитни** | $p < 0.05$ | **Ура!** Различия значимы! |
| | | $p > 0.05$ | Различий нет, всё случайно. |

> **Запомни**: В Шапиро-Уилке мы радуемся **большому p**, а во всех остальных тестах — **маленькому p**.


# Тема 2.3.1: Визуальная проверка на нормальность (QQ-Plot)

Прежде чем доверять математическим тестам (типа Шапиро-Уилка), статистик всегда смотрит на данные глазами.

## 1. Гистограмма vs QQ-Plot
*   **Гистограмма**: показывает "форму" распределения. Мы ищем симметричный колокол.
*   **QQ-Plot (Quantile-Quantile Plot)**: сопоставляет квантили наших данных с квантилями идеального нормального распределения.

## 2. Как читать QQ-Plot?
Представьте прямую линию — это "путь идеала". Точки — это наши реальные данные.
1.  **Точки лежат на линии**: Данные распределены нормально.
2.  **Дуга (изгиб вверх или вниз)**: Распределение скошено (асимметрия). Данные прижаты к одному из краев.
3.  **S-образный изгиб**: У распределения слишком "тяжелые" (длинные) или слишком "легкие" (короткие) хвосты.
4.  **Точки далеко от линии на краях**: Наличие выбросов.

## 3. Зачем это нужно?
T-тест "прощает" небольшие отклонения от линии, если выборка большая. Но если QQ-Plot показывает сильный изгиб или точки-одиночки далеко от линии, это сигнал: **"Используй непараметрику (Манна-Уитни)!"**


````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# 1. Генерируем данные (можно заменить на свои)
data = np.random.normal(loc=0, scale=1, size=100) # Идеально нормальные
# data = np.random.exponential(scale=2, size=100)  # Сильно скошенные (не нормальные)

# Создаем область для двух графиков
plt.figure(figsize=(12, 5))

# ГРАФИК 1: Гистограмма
plt.subplot(1, 2, 1)
plt.hist(data, bins=15, color='skyblue', edgecolor='black', density=True)
plt.title('Гистограмма данных')
plt.xlabel('Значения')
plt.ylabel('Плотность')

# ГРАФИК 2: QQ-Plot
plt.subplot(1, 2, 2)
stats.probplot(data, dist="norm", plot=plt)
plt.title('QQ-Plot (сравнение с нормальным)')

plt.tight_layout()
plt.show()

# Дополнительно выведем тест Шапиро-Уилка для точности
stat, p_val = stats.shapiro(data)
print(f"Тест Шапиро-Уилка: p-value = {p_val:.5f}")

# Твоё задание для практики в PyCharm:
# Запусти код как есть (с экспоненциальными данными). Посмотри, как точки на QQ-Plot выгибаются дугой.
# Закомментируй строку с exponential и раскомментируй строку с normal. Запусти снова. Увидишь, как точки выстроятся в идеальную линию.
````

# 📊 Тема 2.4: Однофакторный дисперсионный анализ (ANOVA)

## 1. Назначение
Используется для сравнения средних значений в **трех и более** группах.
* **Почему не t-тест?** При множественном сравнении групп парами растет вероятность **ошибки I рода** (мы можем найти различия там, где их нет). ANOVA удерживает эту ошибку на уровне 5%.

## 2. Переменные
* **Независимая (Фактор)**: Номинативная (группы/категории).
* **Зависимая**: Количественная (то, что измеряем).

## 3. Логика ANOVA: Разделение изменчивости (SS)
ANOVA делит общую сумму квадратов отклонений ($SS_{total}$) на две части:
1. **$SS_{between}$ (Межгрупповая)**: Показывает разброс **между** средними значениями групп. Это наш "Эффект".
2. **$SS_{within}$ (Внутригрупповая)**: Показывает разброс **внутри** каждой группы. Это наш "Шум".

## 4. Расчет F-значения (Критерий Фишера)
Чтобы получить F, мы должны нормировать суммы квадратов на их степени свободы ($df$):
1. **$df_{between} = m - 1$** (где $m$ — число групп).
2. **$df_{within} = N - m$** (где $N$ — общее число наблюдений).

**Средние квадраты (Mean Squares):**
$$MS_{between} = \frac{SS_{between}}{df_{between}} \quad | \quad MS_{within} = \frac{SS_{within}}{df_{within}}$$

**Итоговое F:**
$$F = \frac{MS_{between}}{MS_{within}} = \frac{\text{Эффект}}{\text{Шум}}$$

## 5. Интерпретация результатов
* **Большое F** $\rightarrow$ **Маленькое p-value** ($p < 0.05$): Отклоняем $H_0$. Хотя бы одна пара средних значимо различается.
* **F близко к 1** $\rightarrow$ **Большое p-value** ($p > 0.05$): Не отклоняем $H_0$. Различия между группами не превышают обычный случайный шум.

## 6. Условия применения
1. Нормальное распределение признака в каждой из групп.
2. Гомогенность (равенство) дисперсий в группах.


## 📝 Практический пример расчета (Stepik)

Представим 3 группы испытуемых по 3 человека в каждой:
*   **Группа 1:** 3, 1, 2 (Среднее $\bar{x}_1 = 2$)
*   **Группа 2:** 5, 3, 4 (Среднее $\bar{x}_2 = 4$)
*   **Группа 3:** 7, 6, 5 (Среднее $\bar{x}_3 = 6$)

### 1. Общее среднее ($\bar{X}_{total}$)
Складываем все числа (сумма = 36) и делим на их количество (9):
$$\bar{X}_{total} = 36 / 9 = 4$$

### 2. Суммы квадратов (SS)
*   **$SS_{total}$ (Общая изменчивость)**: Разница каждого числа и общего среднего (4).
    $SS_{total} = (3-4)^2 + (1-4)^2 + ... + (5-4)^2 = \mathbf{30}$
    $df_{total} = 9 - 1 = 8$
*   **$SS_{between}$ (Эффект групп)**: Насколько средние групп (2, 4, 6) далеки от общего среднего (4).
    $SS_{between} = 3 \cdot (2-4)^2 + 3 \cdot (4-4)^2 + 3 \cdot (6-4)^2 = \mathbf{24}$
    $df_{between} = 3 - 1 = 2$
*   **$SS_{within}$ (Шум/Ошибка)**: Разница каждого числа и среднего **его собственной** группы.
    $SS_{within} = 30 - 24 = \mathbf{6}$
    $df_{within} = 9 - 3 = 6$

### 3. Итоговый расчет F-значения
1.  **$MS_{between}$** $= 24 / 2 = 12$
2.  **$MS_{within}$** $= 6 / 6 = 1$
3.  **$F = 12 / 1 = 12$**

**Вывод:** Наше $F=12$. Это гораздо больше 1, значит межгрупповые различия в 12 раз сильнее случайного шума. Вероятность получить такой результат случайно ($p$) будет очень низкой — мы **отклоняем $H_0$**.


````python
# Задача найти разлия в примении 4 различных типов терапии для корректировки работы дефективного гена

import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import os

# 1. Указываем путь к папке python на рабочем столе Мака
# ~ это твоя домашняя папка, Desktop — рабочий стол, python — твоя папка
file_path = os.path.expanduser('~/Desktop/Python/genetherapy.csv')

# 2. Читаем файл
try:
    df = pd.read_csv(file_path)
    print("✅ Данные из папки Desktop/Python успешно загружены!\n")

    # 3. Проводим ANOVA
    # Модель: результат (expr) зависит от фактора (Therapy)
    model = ols('expr ~ Therapy', data=df).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)

    # 4. Выводим таблицу результатов
    print("📊 Результаты дисперсионного анализа:")
    print(anova_table)

    # 5. Краткая интерпретация
    p_val = anova_table.loc['Therapy', 'PR(>F)']
    if p_val < 0.05:
        print(f"\n🔥 Вердикт: p = {p_val:.6f}. Различия между терапиями ЗНАЧИМЫ!")
    else:
        print(f"\n❄️ Вердикт: p = {p_val:.6f}. Различия не подтверждены.")

except FileNotFoundError:
    print(f"❌ Файл не найден по пути: {file_path}")
    print("Проверь, что папка называется именно 'Python' и файл в ней.")
````

# 🔍 Тема 2.5: Множественные сравнения (Post-hoc тесты)

## 1. Проблема множественных сравнений
Когда ANOVA показывает $p < 0.05$, мы понимаем, что "хотя бы одна пара групп различается".
* **Ловушка**: Если мы начнем сравнивать группы между собой обычными t-тестами (А-B, A-C, A-D, B-C...), шанс совершить **ошибку I рода** (ложноположительный результат) резко растет.
* При 3-х группах риск ошибки уже ~14%, а при 10 группах — более 60%. Это называется **эффектом "рыбалки"**: если закинуть много удочек, какая-то точно поймает "лже-эффект".

## 2. Поправка Бонферрони
Самый простой и жесткий способ борьбы с ошибками.
* **Суть**: Мы делим наш порог значимости ($\alpha = 0.05$) на количество сравнений.
* *Пример*: Если у нас 3 сравнения, то новым порогом значимости станет $0.05 / 3 = 0.016$. Теперь t-тест признается значимым, только если его p-value меньше этого крошечного числа.
* **Минус**: Очень консервативный метод — он так сильно "закручивает гайки", что мы можем пропустить реальные различия.

## 3. Критерий Тьюки (Tukey HSD)
Это золотой стандарт для ANOVA (HSD = *Honestly Significant Difference*).
* **Суть**: Он более умный, чем Бонферрони. Тьюки создает один "критический порог" для разности средних. Если разница между двумя группами больше этого порога — они различаются значимо.
* **Результат**: Мы получаем таблицу, где сравнены **все возможные пары** групп.

## 4. Интерпретация (Что искать в таблице Тьюки)
1. **p-adj (adjusted p-value)**: Это уже "исправленный" p-value. Если он $< 0.05$, значит, эта конкретная пара групп различается значимо.
2. **Доверительный интервал**: Если интервал разности **не включает 0**, значит, между группами есть реальная разница.


### ⚠️ Проблема множественных сравнений и Ошибка I рода

**Ошибка I рода (Ложная тревога)** — это когда мы находим различия там, где их на самом деле нет (отвергаем верную нулевую гипотезу).

#### 🧐 Почему нельзя делать много t-тестов подряд?
Представьте, что вы хотите доказать, что вы экстрасенс, угадывая сторону монетки:
*   Если вы подбросите монетку **1 раз** и угадаете — это может быть случайностью (шанс 5%).
*   Но если вы подбросите монетку **100 раз**, то чисто по закону больших чисел у вас *обязательно* случится серия из нескольких удачных попаданий подряд.
*   Если вы покажете другу только эту серию, вы его обманете. Это и есть **ошибка I рода**.

#### 📈 Накопление ошибки
В статистике мы согласны на 5% риск ошибки ($\alpha = 0.05$). Но при сравнении нескольких групп этот риск **суммируется**:
*   **1 сравнение**: риск ошибки 5%.
*   **10 сравнений** (например, между 5 группами): риск найти хотя бы одно "лже-различие" возрастает до **~40%**!

#### 🛠 Решение: Поправки
Чтобы не стать "лже-экстрасенсом", мы применяем поправки (Бонферрони, Тьюки). Они **"закручивают гайки"**:
1. Снижают порог значимости для каждого отдельного сравнения.
2. Гарантируют, что суммарный риск ошибки во всём эксперименте останется на уровне **5%**.

**Итог:** Поправки нужны, чтобы защитить нас от "ложной тревоги" (ошибки I рода), вероятность которой резко растет при увеличении количества групп.


````python
# Задача найти влияние в применении 4 различных типов терапии для корректировки работы дефективного гена + добавим расчет критерия ТЬЮКИ


import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import os
from statsmodels.stats.multicomp import pairwise_tukeyhsd
# 1. Указываем путь к папке python на рабочем столе Мака
# ~ это твоя домашняя папка, Desktop — рабочий стол, python — твоя папка
file_path = os.path.expanduser('~/Desktop/Python/genetherapy.csv')

# 2. Читаем файл
try:
    df = pd.read_csv(file_path)
    print("✅ Данные из папки Desktop/Python успешно загружены!\n")

    # 3. Проводим ANOVA
    # Модель: результат (expr) зависит от фактора (Therapy)
    model = ols('expr ~ Therapy', data=df).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)

    # 4. Выводим таблицу результатов
    print("📊 Результаты дисперсионного анализа:")
    print(anova_table)

    # 5. Краткая интерпретация
    p_val = anova_table.loc['Therapy', 'PR(>F)']
    if p_val < 0.05:
        print(f"\n🔥 Вердикт: p = {p_val:.6f}. Различия между терапиями ЗНАЧИМЫ!")
    else:
        print(f"\n❄️ Вердикт: p = {p_val:.6f}. Различия не подтверждены.")

except FileNotFoundError:
    print(f"❌ Файл не найден по пути: {file_path}")
    print("Проверь, что папка называется именно 'Python' и файл в ней.")



# Проводим тест Тьюки
# endog — количественные данные (expr)
# groups — группирующая переменная (Therapy)
# alpha — уровень значимости (обычно 0.05)
tukey = pairwise_tukeyhsd(endog=df['expr'], groups=df['Therapy'], alpha=0.05)

print("\n🔍 Результаты теста Тьюки (Сравнение всех пар):")
print(tukey)

# Можно даже визуализировать это (нарисует доверительные интервалы)
tukey.plot_simultaneous()
````

# 📊 Тема 2.6: Многофакторный дисперсионный анализ (Factorial ANOVA)

## 🎯 1. В чем главная суть?
Если обычный ANOVA сравнивает группы по одному признаку (например, только тип терапии), то **многофакторный** позволяет изучать влияние **нескольких факторов одновременно**.

Это позволяет нам увидеть не только влияние каждого фактора в отдельности (**Главные эффекты**), но и их «командную работу» (**Взаимодействие**).

---

## 🤝 2. Взаимодействие факторов (Interaction) — "Сердце" темы
Это ситуация, когда эффект одного фактора **зависит** от того, на каком уровне находится другой фактор.

*   **Пример «Смотря для кого»**:
    *   Лекарство А отлично помогает мужчинам (эффект есть).
    *   То же лекарство А почти бесполезно для женщин (эффекта нет).
    *   **Вывод**: Эффект лекарства **взаимодействует** с полом пациента. Мы не можем сделать общий вывод, не уточнив пол.

---

## 📈 3. Визуализация: Как читать графики взаимодействия?
Мы смотрим на линии, соединяющие средние значения групп:

1.  **Линии параллельны** — Взаимодействия **НЕТ**. Факторы влияют на результат независимо. (Например: женщины всегда ставят оценки выше, чем мужчины, независимо от модели телефона).
2.  **Линии ПЕРЕСЕКАЮТСЯ или НЕ параллельны** — Взаимодействие **ЕСТЬ**. Это признак того, что факторы вступили в "химическую реакцию". Эффект одного фактора «ломается» или меняет направление под влиянием другого.

---

## 🛠 4. Требования к данным (Check-list)
ANOVA — точный инструмент, требующий соблюдения правил:
*   ✅ **Нормальность**: Распределение данных в каждой ячейке должно быть похоже на "колокол".
*   ✅ **Гомогенность дисперсий**: Разброс (шум) внутри всех групп должен быть примерно одинаковым.
*   ✅ **Отсутствие выбросов**: "Супермены" в данных раздувают дисперсию и делают результат ложным.
*   ✅ **Сбалансированность**: Желательно иметь равное количество человек в каждой группе.

---

## ⚖️ 5. Интерпретация результатов (Таблица ANOVA)
В итоговой таблице мы проверяем сразу три гипотезы (смотрим на 3 значения **p-value**):
1.  **Main Effect A**: Влияет ли Фактор А сам по себе?
2.  **Main Effect B**: Влияет ли Фактор B сам по себе?
3.  **Interaction A:B**: Есть ли взаимодействие?
> **Важно**: Если взаимодействие значимо ($p < 0.05$), главные эффекты часто теряют смысл, так как их нужно рассматривать только в связке.

---

### 🚀 Тема 2.7: А/B тестирование

**A/B тест** — это статистический метод проверки гипотез, где сравниваются две версии (контрольная А и тестовая B).

**Ключевые понятия:**
* **Метрика**: То, что мы хотим улучшить (Conversion rate, ARPU, CTR).
* **Statistical Power (Мощность)**: Способность теста найти разницу, если она есть (защита от Ошибки II рода).
* **Sample Size**: Количество людей, которое нужно прогнать через тест, чтобы результат был достоверным.

**Связь с пройденным материалом:**
A/B тест — это практическое применение **t-теста** или **критерия Манна-Уитни**. Мы используем их, чтобы понять: "Разница в 1% конверсии — это победа нашего дизайна или просто случайное колебание?".


### 👢 Бутстреп (Bootstrap) — магия перевыборок

**Бутстреп** — метод оценки распределения статистики (среднего, медианы и др.) путем многократного извлечения выборок "с возвращением" из имеющихся данных.

**Зачем он аналитику?**
* Позволяет считать доверительные интервалы для **любых** метрик (не только средних).
* Не требует нормальности распределения.
* Идеален для A/B тестов, когда данные "грязные" или их мало.

**Минус**: Требует много вычислительной мощности (но для современных ПК 10 000 итераций — это секунда работы).

# 🔗 Тема 3.1: Понятие корреляции

## 🎯 1. В чем суть?
**Корреляция** — это взаимосвязь между двумя **количественными** переменными. Она отвечает на вопрос: "Правда ли, что при изменении X меняется и Y?".

*   **Положительная (+)**: Переменные растут вместе (чем больше рост, тем больше вес).
*   **Отрицательная (-)**: Одна растет, другая падает (чем больше прогулов, тем ниже балл).
*   **Нулевая (0)**: Связи нет (рост никак не связан с номером телефона).

---

## 📏 2. Коэффициент корреляции Пирсона ($r$)
Это число от **-1** до **1**, которое показывает силу и направление **линейной** связи.


| Значение $r$ | Сила связи |
| :--- | :--- |
| **0** | Связи нет |
| **± 0.1 – 0.3** | Слабая |
| **± 0.3 – 0.7** | Умеренная |
| **± 0.7 – 0.9** | Сильная |
| **± 1.0** | Идеальная (все точки на одной прямой) |

---

## ⚠️ 3. Главные "подвохи" корреляции

### А. Корреляция ≠ Причинность (Causality)
Если продажи мороженого и количество лесных пожаров растут одновременно, это не значит, что мороженое поджигает леса. У них есть общая причина — **жара**.

### Б. Чувствительность к выбросам
Коэффициент Пирсона очень боится "суперменов".
*   **Пример**: У нас есть 10 человек с идеальной корреляцией ($r=0.9$). Добавьте одного человека с огромным ростом и крошечным весом (выброс) — и корреляция может упасть до **0** или даже стать отрицательной.

### В. Только линейные связи
Если твои точки на графике образуют **дугу** или **параболу**, Пирсон покажет $0$, хотя связь на самом деле есть. Просто она нелинейная.

## 📐 5. Ковариация и Коэффициент детерминации ($R^2$)

### А. Ковариация ($cov$) — "Сырая" связь
Это мера того, насколько две переменные меняются вместе.
* **Суть**: Если при отклонении X от своего среднего, Y отклоняется от своего в ту же сторону — ковариация положительная.
* **Проблема**: Она зависит от единиц измерения. Если измерять рост в метрах — будет одно число, в миллиметрах — в тысячу раз больше. Это неудобно.
> **Лайфхак**: Коэффициент корреляции Пирсона — это просто **стандартизированная ковариация**, которая всегда зажата между -1 и 1.


### Внутри ковариации: Как это считается?

**Ковариация ($cov_{xy}$)** — это среднее произведение отклонений каждой пары точек от их средних значений.

#### 📝 Формула:
$$cov_{xy} = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{n - 1}$$

#### 🔍 Как это читать "по-русски":
1. Берем точку, смотрим насколько её **X** далек от среднего $\bar{x}$.
2. Смотрим насколько её **Y** далек от среднего $\bar{y}$.
3. Перемножаем эти отклонения.
4. Повторяем для всех точек и делим на $n-1$.

**Почему это работает?**
* Если X и Y отклоняются в **одну сторону** (оба больше среднего или оба меньше) — произведение будет **положительным** (+ на + или - на -).
* Если они "разбегаются" (один больше, другой меньше) — произведение будет **отрицательным**.

---

### 🧬 Связь Ковариации и Корреляции (Пирсона)

Корреляция Пирсона ($r_{xy}$) — это просто ковариация, которую "укротили", чтобы она не зависела от масштаба (метров, килограммов и т.д.). Для этого её делят на произведение стандартных отклонений:

$$r_{xy} = \frac{cov_{xy}}{sd_x \cdot sd_y}$$

> **Вывод**: Ковариация говорит только о **направлении** (плюс или минус), а Корреляция — еще и о **силе** (насколько близко к 1 или -1).


### Б. Коэффициент детерминации ($R^2$) — "Объясняющая сила"
Это квадрат коэффициента корреляции Пирсона ($r^2$).
* **Что показывает**: Какая доля изменчивости одной переменной (Y) обусловлена (объясняется) влиянием другой переменной (X).
* **Пример**: Если корреляция между временем подготовки и баллом $r = 0.7$, то $R^2 = 0.49$.
* **Интерпретация**: "Мы на **49%** можем объяснить успех на экзамене временем подготовки. Остальные 51% — это другие факторы (удача, интеллект, выспался ли студент)".

# 📐 Математика корреляции: От отклонений к коэффициенту

Чтобы понять, как две переменные "общаются" между собой, статистика проходит три этапа расчетов.

### 1. Произведение отклонений (The Core)
Для каждой пары точек мы смотрим, в какой "квадрат" относительно центра они попадают:
$$(x_i - \bar{x}) \cdot (y_i - \bar{y})$$
*   **Плюс (+) на Плюс (+)** или **Минус (-) на Минус (-)** — точка подтверждает **положительную** связь (лежит в правом верхнем или левом нижнем углу).
*   **Плюс на Минус** — точка указывает на **отрицательную** связь.

---

### 2. Ковариация ($cov_{xy}$) — "Сырая" связь
Это средняя сумма всех произведений отклонений. Она показывает **направление** связи, но её масштаб зависит от единиц измерения (метры, граммы и т.д.).
$$cov_{xy} = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{N - 1}$$

---

### 3. Коэффициент корреляции Пирсона ($r_{xy}$)
Чтобы избавиться от влияния масштаба, мы делим ковариацию на произведение стандартных отклонений ($\sigma$). Это превращает "сырое" число в универсальный показатель от **-1 до +1**.

$$r_{xy} = \frac{cov_{xy}}{\sigma_x \cdot \sigma_y}$$

---

## 🐍 Проверяем математику кодом:
Если мы захотим посчитать это "вручную" (без готовой функции `pearsonr`), код будет выглядеть так:

```python
import numpy as np

# 1. Считаем ковариацию (берем значение из матрицы 2x2)
cov_xy = np.cov(x, y)[0, 1]

# 2. Считаем стандартные отклонения
sd_x = np.std(x, ddof=1)
sd_y = np.std(y, ddof=1)

# 3. Финальный r
r_custom = cov_xy / (sd_x * sd_y)
print(f"Рассчитанный r: {r_custom:.4f}")
```
> **Важно**: В жизни мы используем `scipy.stats.pearsonr`, но знание этой формулы помогает понять, почему выбросы (большие отклонения) так сильно "ломают" итоговый результат.

````python
import numpy as np
from scipy import stats

# Создадим две группы для примеров
group_a = np.array([10, 12, 11, 13, 12, 11, 40]) # Группа с выбросом (40)
group_b = np.array([15, 14, 16, 15, 14, 15, 16])

print("--- 1. ОПИСАТЕЛЬНАЯ СТАТИСТИКА ---")
print(f"Z-score (стандартизация): {stats.zscore(group_a)}")
print(f"Межквартильный размах (IQR): {stats.iqr(group_a)}")

print("\n--- 2. ПРОВЕРКА УСЛОВИЙ ---")
# Проверка на нормальность (H0: данные нормальные)
shapiro_p = stats.shapiro(group_a).pvalue
print(f"Шапиро-Уилк p-value: {shapiro_p:.4f} ({'Нормально' if shapiro_p > 0.05 else 'Криво'})")

# Проверка равенства дисперсий (H0: дисперсии равны)
levene_p = stats.levene(group_a, group_b).pvalue
print(f"Тест Левена p-value: {levene_p:.4f}")

print("\n--- 3. СРАВНЕНИЕ ГРУПП ---")
# T-тест (если всё нормально)
t_stat, t_p = stats.ttest_ind(group_a, group_b)
print(f"T-тест: t = {t_stat:.2f}, p = {t_p:.4f}")

# Манн-Уитни (если есть выбросы или ненормальность)
u_stat, u_p = stats.mannwhitneyu(group_a, group_b)
print(f"Манн-Уитни: u = {u_stat:.2f}, p = {u_p:.4f}")

print("\n--- 4. КОРРЕЛЯЦИЯ ---")
# Пирсон (линейная связь)
r_p, p_p = stats.pearsonr(group_a[:5], group_b[:5])
print(f"Пирсон: r = {r_p:.2f}, p = {p_p:.4f}")

# Спирмен (если связь кривая или есть выбросы)
r_s, p_s = stats.spearmanr(group_a, group_b)
print(f"Спирмен: rho = {r_s:.2f}, p = {p_s:.4f}")
````

# 🛡️ Тема 3.2: Условия применения корреляции и "План Б"

Коэффициент Пирсона ($r$) — мощный инструмент, но он работает только в идеальных условиях. Если данные "грязные", нам нужен другой подход.

## ⚠️ 1. Correlation is NOT Causation (Корреляция ≠ Причинность)
Самая частая ошибка аналитиков. Наличие связи между X и Y не означает, что X вызывает Y.
*   **Третья переменная**: Продажи мороженого и нападения акул коррелируют, потому что оба зависят от **жары** (скрытый фактор).
*   **Обратная связь**: Мы не всегда знаем, что причина, а что следствие.
> **Вывод**: Корреляция лишь повод для гипотезы, а не доказательство вины.

---

## ⚠️ 2. Когда Пирсон показывает 0?
Если $r \approx 0$, это **не всегда** значит, что связи нет. Пирсон "слепнет" в двух случаях:
1.  **Связи действительно нет**: Точки разбросаны как хаотичное облако.
2.  **Связь НЕЛИНЕЙНАЯ**: Точки образуют дугу, круг или параболу.
    *   Пирсон ищет только **прямые линии**. Если вы видите на графике "радугу", Пирсон выдаст 0, хотя связь там идеальная.

---

## 🛠️ 3. Коэффициент Спирмена ($\rho$) — наш "План Б"
Когда условия для Пирсона (линейность и нормальность) нарушены, на сцену выходит **Корреляция Спирмена**.

### В чем его суперсила?
*   Он работает не с самими числами, а с их **рангами** (местами в очереди).
*   **Устойчив к выбросам**: "Супермен" с баллом 1000 просто получит 1-е место, и его гигантское значение не "раздует" формулу.
*   **Видит нелинейные связи**: Если X растет и Y растет (пусть и по дуге), Спирмен это заметит и покажет высокую корреляцию.

---

# 📈 Тема 3.3. Регрессия с одной независимой переменной (простая линейная регрессия)

## 1. Что такое простая линейная регрессия?

Метод для моделирования **линейной связи** между двумя количественными переменными:

- **Зависимая переменная (y)** – отклик, предсказываемая величина.
- **Независимая переменная (x)** – предиктор, объясняющая переменная.

Уравнение регрессии:

\[
y = \beta_0 + \beta_1 \cdot x + \varepsilon
\]

- \(\beta_0\) – **свободный член (intercept)**: значение \(y\) при \(x = 0\).
- \(\beta_1\) – **коэффициент наклона (slope)**: на сколько в среднем изменяется \(y\) при увеличении \(x\) на 1.
- \(\varepsilon\) – случайная ошибка, предполагается, что её среднее равно 0.

## 2. Как оцениваются коэффициенты $\beta_0$ и $\beta_1$?

Используется **метод наименьших квадратов (Ordinary Least Squares, OLS)**.  

**Идея:** найти такие $\hat{\beta}_0$ и $\hat{\beta}_1$, чтобы сумма квадратов остатков была минимальной.

### Формулы расчета:

$$
\hat{\beta}_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2} = \frac{\text{Cov}(x,y)}{\text{Var}(x)}
$$

$$
\hat{\beta}_0 = \bar{y} - \hat{\beta}_1 \cdot \bar{x}
$$

**Где:**
* $\bar{x}, \bar{y}$ — выборочные средние.
* $\text{Cov}(x,y)$ — ковариация.
* $\text{Var}(x)$ — дисперсия $x$.

> **Важно:** эти формулы всегда дают единственное решение, если в данных есть вариация по $x$ (не все значения одинаковы).## 2. Как оцениваются коэффициенты $\beta_0$ и $\beta_1$?

Используется **метод наименьших квадратов (Ordinary Least Squares, OLS)**.  

**Идея:** найти такие $\hat{\beta}_0$ и $\hat{\beta}_1$, чтобы сумма квадратов остатков была минимальной.

### Формулы расчета:

$$
\hat{\beta}_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2} = \frac{\text{Cov}(x,y)}{\text{Var}(x)}
$$

$$
\hat{\beta}_0 = \bar{y} - \hat{\beta}_1 \cdot \bar{x}
$$

**Где:**

> **Важно:** эти формулы всегда дают единственное решение, если в данных есть вариация по $x$ (не все значения одинаковы).
## 3. Интерпретация коэффициентов

- **\(\beta_1\) (slope)**:
  - Положительный → с ростом \(x\) переменная \(y\) в среднем увеличивается.
  - Отрицательный → с ростом \(x\) переменная \(y\) в среднем уменьшается.
  - Ноль → линейной связи нет.
- **\(\beta_0\) (intercept)**:
  - Имеет смысл только если \(x = 0\) входит в область допустимых значений. Иначе это просто «сдвиг» модели.

# 📊 Тема 3.4. Гипотеза о значимости взаимосвязи и коэффициент детерминации

## 1. Зачем проверять значимость регрессии?

После того как мы построили уравнение регрессии $y = \beta_0 + \beta_1 x$, возникает вопрос: **а есть ли вообще линейная связь между $x$ и $y$?** Может быть, полученный наклон $\hat{\beta}_1$ отличается от нуля просто из-за случайных колебаний в данных?

Для ответа на этот вопрос используется **статистическая проверка гипотез**.

---

## 2. Гипотезы

* **$H_0: \beta_1 = 0$** — Линейной зависимости нет (наклон в генеральной совокупности равен нулю).
* **$H_1: \beta_1 \neq 0$** — Существует статистически значимая линейная зависимость.

Если мы отвергаем $H_0$, значит, связь **значима**.

---

## 3. t-тест для коэффициента наклона $\beta_1$

Для проверки гипотезы вычисляется **t-статистика**:

$$
t = \frac{\hat{\beta}_1}{SE(\hat{\beta}_1)}
$$

Где $SE(\hat{\beta}_1)$ — стандартная ошибка наклона. Она показывает вариативность коэффициента.

* **Степени свободы:** $df = n - 2$.
* **Правило принятия решения:**
    * Если $p\text{-value} < \alpha$ (обычно $0.05$), отвергаем $H_0$ — **связь значима**.
    * Если $p\text{-value} \ge \alpha$, **нет оснований** считать связь значимой.

---

## 4. Коэффициент детерминации $R^2$

Показывает, **какую долю дисперсии** зависимой переменной $y$ объясняет наша модель.

$$
R^2 = \frac{SS_{reg}}{SS_{total}} = 1 - \frac{SS_{res}}{SS_{total}}
$$

**Компоненты:**
* $SS_{total} = \sum (y_i - \bar{y})^2$ — Общая сумма квадратов (общая изменчивость).
* $SS_{reg} = \sum (\hat{y}_i - \bar{y})^2$ — Объясненная сумма квадратов (изменчивость модели).
* $SS_{res} = \sum (y_i - \hat{y}_i)^2$ — Остаточная сумма квадратов (ошибки).

> **Важно:** В парной регрессии $R^2$ равен квадрату коэффициента корреляции Пирсона ($r^2$).

---

## 5. Расчет в Python (scipy.stats)

Библиотека `scipy` позволяет получить все эти параметры одной функцией:

# 📊 Тема 3.5. Условия применения простой линейной регрессии

Чтобы результаты регрессионного анализа были достоверными, а выводы — статистически значимыми, данные должны соответствовать ряду условий (ассумпций).

---

## 1. Линейность (Linearity)
Связь между независимой переменной $X$ и зависимой переменной $Y$ должна быть **линейной**.
* **Как проверить:** Визуально с помощью графика рассеяния (*Scatter plot*). Точки должны выстраиваться вдоль воображаемой прямой, а не дуги или кривой.
* **Что делать, если нарушено:** Использовать преобразования переменных (например, логарифмирование) или нелинейную регрессию.

---

## 2. Независимость остатков (Independence)
Ошибки (остатки) не должны быть зависимы друг от друга.
* Это условие особенно важно для временных рядов, где текущее значение часто зависит от предыдущего (автокорреляция).
* **Как проверить:** Тест Дарбина-Уотсона.

---

## 3. Гомоскедастичность (Homoscedasticity)
Дисперсия остатков должна быть постоянной на всех уровнях значения предиктора $X$.
* **Простыми словами:** Разброс точек вокруг линии регрессии должен быть примерно одинаковым в начале, середине и конце графика.
* **Как проверить:** График остатков (*Residuals vs Fitted*). Если график напоминает "рупор" или "факел" (разброс расширяется или сужается), это **гетероскедастичность**.

---

## 4. Нормальное распределение остатков (Normality of Residuals)
Ошибки модели должны быть распределены нормально со средним значением, близким к нулю.
* Это критически важно для корректного расчета $p-value$ и доверительных интервалов.
* **Как проверить:** * Визуально: гистограмма остатков или график **Q-Q plot** (точки должны лежать на диагональной линии).
    * Статистически: тест Шапиро-Уилка или тест Колмогорова-Смирнова.

---

## 5. Отсутствие выбросов (Outliers)
Линейная регрессия очень чувствительна к экстремальным значениям, так как метод наименьших квадратов (OLS) минимизирует *квадраты* отклонений. Одно выбивающееся значение может сильно "притянуть" линию к себе.
* **Как проверить:** График "ящик с усами" (*Boxplot*) или расчет расстояния Кука (*Cook's distance*).

---

## 💡 Резюме для запоминания (Checklist):
1. **L**inearity — Линейность.
2. **I**ndependence — Независимость.
3. **N**ormality — Нормальность остатков.
4. **E**qual Variance — Гомоскедастичность.

> **Важно:** Если условия нарушены, оценки коэффициентов $\hat{\beta}_0$ и $\hat{\beta}_1$ могут остаться несмещенными, но их стандартные ошибки будут рассчитаны неверно, что приведет к ложным выводам о значимости модели.

# 🧪 3.6. Применение регрессионного анализа и интерпретация результатов

## 📌 Контекст исследования
Исследуется взаимосвязь между уровнем бедности и уровнем образования в штатах США.  
**Данные:** 51 наблюдение (50 штатов + округ Колумбия).

### 📊 Переменные (states.csv):
* `hs_grad` — доля населения со средним образованием (%) — **предиктор (X)**.
* `poverty` — уровень бедности (%) — **зависимая переменная (Y)**.

---

## 🎯 Цель работы
Построить модель простой линейной регрессии, оценить её параметры и проверить статистическую значимость связи.

---

## 📋 План анализа и требования к результатам

### 1. Описательная статистика
Необходимо рассчитать ключевые показатели для переменных `hs_grad` и `poverty`:
* Среднее ($\bar{x}, \bar{y}$), стандартное отклонение ($sd$), минимум и максимум.

### 2. Модель регрессии
Оценка коэффициентов уравнения:
$$\text{poverty} = \hat{\beta}_0 + \hat{\beta}_1 \cdot \text{hs\_grad}$$

**Ожидаемые результаты (согласно тестовым данным):**
* $\hat{\beta}_0$ (Intercept) $\approx 64.78$
* $\hat{\beta}_1$ (Slope) $\approx -0.62$

### 3. Коэффициент детерминации $R^2$
Показывает долю объясненной дисперсии зависимой переменной.
* **Ожидаемое значение:** $R^2 \approx 0.558$ (модель объясняет ~56% изменчивости уровня бедности).

### 4. Таблица коэффициентов
Вывод полной статистики по параметрам модели:

| Variable | Estimate | Std. Error | t value | $Pr(>|t|)$ |
| :--- | :--- | :--- | :--- | :--- |
| **Intercept** | 64.7810 | 6.8026 | 9.52 | < 0.001 |
| **hs_grad** | -0.6212 | 0.0790 | -7.86 | < 0.001 |

````python
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
import pandas as pd



file_path = 'http://d396qusza40orc.cloudfront.net/statistics/lec_resources/states.csv'

df = pd.read_csv(file_path)
X = df[['hs_grad']]
Y = df[['poverty']]

X_const = sm.add_constant(X)
model = sm.OLS(Y, X_const).fit()
print(model.summary())

print('*'*100)
print()


x = df['hs_grad']
y = df['poverty']

res = stats.linregress(x, y)

print(f"slope = {res.slope:.4f}")
print(f"intercept = {res.intercept:.4f}")
print(f"stderr = {res.stderr:.4f}")


plt.scatter(x, y, label = 'Данные', color = 'red')
plt.plot(x, res.intercept + x * res.slope, color ='blue', label='Линия регрессии')
plt.xlabel('Уровень образования (%)')
plt.ylabel('Уровень бедности (%)')
plt.title(f'Зависимость уровня бедности от образования\nR² = {res.rvalue**2:.3f}, p = {res.pvalue:.3f}')
plt.text(x=78, y=6, s=f'x = {res.slope:.2f}·x + {res.intercept:.2f}')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()



print("\nДоверительные интервалы коэффициентов (95%):")
print(model.conf_int())


print('*'*100)


residuals = model.resid
shapiro_stat, shapiro_p = stats.shapiro(residuals)
print(f"\nПроверка нормальности остатков (Shapiro–Wilk): p = {shapiro_p:.4f}")
if shapiro_p > 0.05:
    print("Остатки распределены нормально (условие выполнено).")
else:
    print("Остатки не нормальны (возможны проблемы с выводами).")



print()

print('*'*100)


new_x = 85
new_x_const = sm.add_constant([[1, new_x]])  # константа и значение
predicted = model.predict(new_x_const)[0]
print(f"При hs_grad = {new_x}% прогнозируемый poverty = {predicted:.2f}%")




print()


print('*'*100)
results_summary = pd.DataFrame({
    'Показатель': ['R²', 'F-статистика', 'p-value (модель)', 'Уравнение'],
    'Значение': [f"{res.rvalue**2:.3f}", f"{model.fvalue:.2f}", f"{model.f_pvalue:.2e}",
                 f"poverty = {res.slope:.2f}·hs_grad + {res.intercept:.2f}"]
})
print("\nКраткая сводка модели:")
print(results_summary.to_string(index=False))

````
